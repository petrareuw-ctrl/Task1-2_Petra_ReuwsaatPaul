You are analyzing neural recordings from a population of 50 neurons in visual cortex. Each neuron has a preferred orientation (the visual angle it responds to most strongly). Your goal is to understand how a group of these neurons works together in the presence of noise.

Create a population of 50 neurons where each neuron has a randomly assigned preferred orientation between 0Â° and 180Â°. [hint: use np.random.uniform()]

```python
n_neurons=50
neurons=np.random.uniform(0, 180, n_neurons)


```

## 2: Simulate Neural Responses
Check the neurons responses to 8 different stimulus orientations: 0Â°, 22.5Â°, 45Â°, 67.5Â°, 90Â°, 112.5Â°, 135Â°, and 157.5Â°. write a function to calculate neurons response.

Write a function that calculates how strongly a neuron fires:
â€¢ Inputs: stimulus orientation, neuronâ€™s preferred orientation and maximum firing rate (r_max = 100 spikes/second)
â€¢ Output: the firing rate (response)
â€¢ Your function should:
â€¢ Calculate the difference between the stimulus angle and the neuron's preferred angle
â€¢ Remember: orientations are circular! 0Â° and 180Â° are similar, remember to handle wraparound [hint: use np.minimum(diff, 180 - diff)]
â€¢ Use cosine tuning: response = r_max * cos(difference in radians)
â€¢ Make sure responses are not negative

```python
def strengh(stimulus, prefered, r_max=100):
    dif = abs(stimulus - prefered)
    dif = np.minimum(dif, 180 - dif)
    dif = np.radians(dif)
    response = r_max*np.cos(dif)
    return max(0, response)
```

## 3. Generate response data
Simulate responses to 8 stimulus orientations: 0Â°, 22.5Â°, 45Â°, 67.5Â°, 90Â°, 112.5Â°, 135Â°, and 157.5Â°. 
For each combination of neuron and stimulus:
â€¢ Calculate the response
â€¢ Generate 30 trials with Poisson noise(*)
â€¢ Store the results in a 3D array with shape (n_neurons, n_stimuli, n_trials)
â€¢ Save results to file, we will use it in the next task ðŸ˜Š

(*) Why Poisson noise? Real neurons don't fire exactly the same number of spikes every time - there's randomness. Poisson noise captures this biological variability.

```python
stimulus = [0, 22.5, 45, 67.5, 90, 112.5, 135, 157.5]

result = np.zeros((n_neurons, len(stimulus), 30))

for i in range(n_neurons):
    for j in range(len(stimulus)):
        for k in range(30):
            result[i, j, k] = np.random.poisson(strengh(stimulus[j], neurons[i]))
```

```python
result
```

```python
np.savez_compressed('result.npz', 
                    result=result,
                    neurons=neurons,
                    stimulus=stimulus)

data = np.load("result.npz")
result_loaded = data["result"]
result_loaded

```
